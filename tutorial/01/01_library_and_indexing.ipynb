- 데이터프레임 : 엑셀처럼 행/열 구조를 갖춘 데이터 자료형

인덱싱이란 앞에서 봤던 1차원 리스트나 2차원 데이터 프레임 같은 데이터 덩어리로부터 특정 데이터만 추출하는 행위를 말합니다. 인덱싱 방법은 매우 다양하며 일반적으로 쓰이는 방법들을 소개해 드리겠습니다.

### 2-1. 열 인덱싱

데이터 프레임은 행과 열로 이루어진 2차원 자료형입니다. 만약 특정 열만 추출하고 싶다면 데이터 프레임 뒤에 대괄호로 컬럼 이름을 써주면 됩니다.

````python
# 특정 열을 모두 추출할 때
df['Close'] 
````

앞에서 언급했지만 df 변수에 담은 데이터 프레임은 이중 컬럼을 가지고 있습니다. Close 컬럼을 인덱싱하게되면 이처럼 Close 컬럼에 포함되어 있는 하위 컬럼(열)들을 추출하게 됩니다.

​여기서 다시 한번 열 인덱싱을 수행하면 다음과 같이 SPY 컬럼에 해당하는 데이터만 뽑아낼 수 있습니다


````python
# 현재 data는 이중 컬럼이 적용되어 있다. 따라서 2 단계 인덱싱이 가능하다.
# 먼저 Close 열을 모두 추출한 다음 연속해서 SPY 열만 뽑아낼 수 있다.
df['Close']['SPY']
````

꼭 하나의 열만 추출할 수 있는 것은 아닙니다. 2개 이상의 컬럼은 대괄호로 감싸서 아래와 같이 추출할 수도 있습니다.

````python
# 만약 2개의 컬럼(열)을 함께 추출하고 싶다면 대괄호를 이중으로 겹쳐서 사용하면 된다.
df['Close'][['SPY','TSLA']]
````

그런데 우리가 종가(Close)만 계속해서 볼 텐데 이렇게 이중으로 인덱싱을 하는 것은 매우 귀찮죠? 이때는 Close 컬럼을 인덱싱한 데이터 프레임을 복사해서 다른 변수에 담아주면 됩니다.

````python
# 우리는 Close 컬럼만 계속 다룰 것이므로 사용하기 편하게 데이터프레임을 다른 변수로 복사하자.
df_close = df['Close'].copy()
df_close
````

여기서는 copy() 함수를 사용했습니다. 해당 함수를 사용하지 않고 그대로 변수에 넣어줄 수도 있습니다. 그러나 그런 방식은 잘 못 사용하면 이따끔 상당히 난처한 상황이 생길 수도 있습니다. 대입한 변수를 수정했을 때 대입한 대상인 원본 변수가 수정될 수 있기 때문입니다. 데이터 분석을 하다 보면 뭔가 처리를 잘못해서 원본 데이터를 다시 가져다가 써야 할 때가 있는데, 원본이 수정되어버리면 처음부터 데이터를 다시 가져와야 합니다.


한번 코드를 보면서 무슨 말인지 눈으로 확인해 보겠습니다.

````python
# 카피를 하지 않고 변수에 그대로 넣어줄 수도 있다.
# 다만, 이럴 경우 새로 만든 변수 내 데이터를 수정하면 원본도 함께 수정된다. 
# 객체 참조라는 개념으로 발생하는 것인데, 
# 새로운 변수에 다른 변수를 그대로 할당하면,
# 메모리상에는 동일한 주소값을 가리키고 있기 때문에 해당 주소의 데이터를 바꾸면 원본 데이터가 바뀌는 것이다.
# 테스트해보자.

# 아래와 같이 새로운 변수에 할당해주었다. 변수에 담긴 값은 동일하다.
df_ref = df_close
# 비교를 위해 copy한 변수도 만들어주겠다.
df_copy = df_close.copy()
````

````python
df_ref
````

잠깐 참고로, 주석(#)에 적힌 객체 참조라는 개념에 대해서는 나중에 한 번은 찾아보시길 권장 드립니다. 프로그래밍이 처음이라면 메모리와 데이터 저장에 대한 개념이 와닿지 않을 수 있지만 알아두면 이러한 현상에 대해 더 이상 마법이라 느껴지지 않을 것입니다.

df_close라는 변수는 기존 데이터 프레임에서 Close 컬럼만 인덱싱한 데이터 프레임을 담고 있습니다. 따라서 df_ref는 해당 데이터 프레임을 그대로 가리키고 있는 꼴이라 생각하시면 되겠습니다. 반면 df_copy는 df_close라는 변수를 복제해서 새로운 데이터 프레임을 만든 것입니다. 

그래서 df_close와 df_ref는 같은 데이터 프레임을 다른 변수명으로 담고 있으며, df_copy는 같은 모양(같은 데이터 값)이지만 실제로는 모양만 같고 다른 데이터 프레임(다른 메모리에 저장되어 있는 데이터)을 담고 있게 됩니다.

그래서 아래와 같이 df_close와 같은 데이터 프레임을 가리키고 있는 df_ref를 수정하면 df_close도 동일하게 수정됩니다.

````python
# 여기서 df_ref를 수정하면 df_close도 바뀐다.
# 아래 명령어는 'FILL_0'라는 새로운 컬럼을 생성하고, 0으로 채워주겠다는 의미다.
df_ref['FILL_0'] = 0
df_ref
````

````python
# df_close는 건드리지도 않았는데 아래와 같이 수정되었다.
df_close
````

반면 다른 메모리 공간에 저장된 다른 데이터 프레임을 가리키고 있는 df_copy의 경우 변하지 않았습니다

````python
# 이를 방지하기 위해서 copy를 사용한다. 
# 분석하다보면 원본을 다시 확인해야 할 때가 많기 때문이다.
# copy를 사용했던 변수는 그대로 원본을 유지하고 있다.
df_copy
````

df_copy를 수정하더라도 마찬가지입니다. drop()을 통해 TSLA 컬럼을 제거한 다음 원본 df_close를 확인하면 그대로 유지되고 있는 것을 볼 수 있습니다. drop() 함수에 대해서는 주석을 참고해 주세요.

````python
# 해당 변수는 데이터를 바꿔도 원본에 영향을 미치지 않는다. 메모리상에 새로운 주소값에 위치해있기 때문이다.

# drop은 특정 행이나 열을 제거하는 함수이며, axis=1옵션은 '열'을 제거하겠다는 의미다. default는 0이며, 행을 제거한다.
# inplace 옵션을 True로 주면, 변경된 값으로 변수를 다시 저장해준다. 
# 만약 해당 변수가 수정되면 안되는 경우 inplace=False 옵션을 주거나 inplace 옵션을 비워두면 된다. default는 False다.
df_copy.drop('TSLA', axis=1, inplace=True)
df_copy
````

````python
# 다시 df_close를 확인해보면 원본이 수정되지 않았다는 것을 알 수 있다.
df_close
````

### 2-3. 행, 열 동시 인덱싱

위에서 확인한 행 인덱싱, 열 인덱싱을 함께 사용할 수도 있습니다.

````python
# 두 방식을 같이 사용하면 loc행과 열을 함께 인덱싱할 수도 있다.
df_close[:10][['SPY','TSLA']]
````

````python
# 순서는 상관없다.
df_close[['SPY','TSLA']][:10]
````

그리고 이 케이스 역시 자체 함수를 제공합니다. 함수는 대표적으로 2개가 있습니다.

먼저 인덱스 이름, 컬럼 이름으로 인덱싱하는 loc() 함수입니다.

````python
# 자체 함수를 사용할 수도 있다.
# 함수는 loc()와 iloc()가 있으며, 
# loc는 location의 약자다. 
# 아래 코드는 첫 행부터 2020-01-13 인덱스가 위치한 행까지, 첫 열부터 'SPY' 컬럼이 위치한 열까지 추출한다.
df_close.loc[:'2020-01-13', :'SPY']
````

````python
# 만약 '~까지'가 아니라 특정 행, 특정 열을 지정하고 싶다면 다음과 같이 작성할 수 있다.
df_close.loc['2020-01-13', 'SPY']
````

````python
# 물론 여러 컬럼, 여러 행을 동시에 추출하는 것도 가능하다.
# 이 때는 대괄호를 한번 더 씌워준다.
df_close.loc[['2020-01-13','2019-12-31'], ['SPY','QQQ']]
````

다음은 n 번째 행, n 번째 열처럼 숫자로 인덱싱하는 iloc() 함수입니다.

````python
# 특정 인덱스와 컬럼 이름 대신 숫자로 위치만 넘겨줄 수도 있다.
# 이 때는 iloc(integer location)을 사용한다.
# 다음 코드는 첫 행부터 10개 행까지, 첫 열부터 2번째 열까지 추출한다.
df_close.iloc[:10, :2]
````

지금까지 라이브러리, 데이터 프레임, 그리고 인덱싱까지 간단하게 살펴보았습니다.

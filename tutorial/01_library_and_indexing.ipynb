## Step 1. Library

- 라이브러리 : 우리가 파이썬을 사용할 때, 모든 코드를 한땀한땀 구현하는 것이 아니라 다른 사람들이 미리 개발해놓은 코드를 가져다 쓸 수 있게 제공하는 것

````python
# yfinance라는 라이브러리를 yf라는 이름으로 사용하겠다.
import yfinance as yf
````

> 참고 : [yfinance.Tickers()](https://github.com/ranaroussi/yfinance/blob/main/yfinance/tickers.py)

````python
# yf에서 제공하는 Tickers 클래스에 인자(입력)를 넣어서 stocks라는 인스턴스를 생성한다. 
# 클래스는 추상적인 코드 형상이라면, 인스턴스는 우리가 다룰 수 있는 '실체'다.
# 예를 들어, 포켓몬 빵은 클래스다. 오늘 집에 오면서 사온 피카츄빵은 인스턴스다. 
# 내가 지금 먹고 있는 것은 포멧몬 빵(클래스) 종류 중에 하나인 피카츄 빵(인스턴스)이다.
stocks = yf.Tickers(['TSLA', 'SPY', 'QQQ'])
````

````python
# 인스턴스 내부 __repr__()함수에 의해 다음과 같이 출력됨.
stocks
````

````python
# history() 함수를 사용해 주가 정보를 가져올 수 있다
# history() 함수는 내부적으로 download() 함수를 호출하며, 해당 함수를 통해 Pandas 데이터프레임을 생성한다.
# 데이터프레임은 판다스에서 제공하는 2차원 자료형으로, 행과 열이 있는 엑셀과 같은 형태를 띈다.
# 변수는 일반적으로 df를 많이 사용하며, DataFrame의 약자다.
df = stocks.history(start='2020-01-01', end='2022-06-30')
df
````

## Step 2. Indexing

- 인덱싱 : 데이터 프레임에서 특정 데이터만 추출하는 것
- 데이터프레임 : 엑셀처럼 행/열 구조를 갖춘 데이터 자료형

### 2-1. 열 인덱싱

````python
# 특정 열을 모두 추출할 때
df['Close'] 
````

````python
# 현재 data는 이중 컬럼이 적용되어 있다. 따라서 2 단계 인덱싱이 가능하다.
# 먼저 Close 열을 모두 추출한 다음 연속해서 SPY 열만 뽑아낼 수 있다.
df['Close']['SPY']
````

````python
# 만약 2개의 컬럼(열)을 함께 추출하고 싶다면 대괄호를 이중으로 겹쳐서 사용하면 된다.
df['Close'][['SPY','TSLA']]
````

````python
# 우리는 Close 컬럼만 계속 다룰 것이므로 사용하기 편하게 데이터프레임을 다른 변수로 복사하자.
df_close = df['Close'].copy()
df_close
````

````python
# 카피를 하지 않고 변수에 그대로 넣어줄 수도 있다.
# 다만, 이럴 경우 새로 만든 변수 내 데이터를 수정하면 원본도 함께 수정된다. 
# 객체 참조라는 개념으로 발생하는 것인데, 
# 새로운 변수에 다른 변수를 그대로 할당하면,
# 메모리상에는 동일한 주소값을 가리키고 있기 때문에 해당 주소의 데이터를 바꾸면 원본 데이터가 바뀌는 것이다.
# 테스트해보자.

# 아래와 같이 새로운 변수에 할당해주었다. 변수에 담긴 값은 동일하다.
df_ref = df_close
# 비교를 위해 copy한 변수도 만들어주겠다.
df_copy = df_close.copy()
````

````python
df_ref
````

````python
# 여기서 df_ref를 수정하면 df_close도 바뀐다.
# 아래 명령어는 'FILL_0'라는 새로운 컬럼을 생성하고, 0으로 채워주겠다는 의미다.
df_ref['FILL_0'] = 0
df_ref
````

````python
# df_close는 건드리지도 않았는데 아래와 같이 수정되었다.
df_close
````

````python
# 이를 방지하기 위해서 copy를 사용한다. 
# 분석하다보면 원본을 다시 확인해야 할 때가 많기 때문이다.
# copy를 사용했던 변수는 그대로 원본을 유지하고 있다.
df_copy
````

````python
# 해당 변수는 데이터를 바꿔도 원본에 영향을 미치지 않는다. 메모리상에 새로운 주소값에 위치해있기 때문이다.

# drop은 특정 행이나 열을 제거하는 함수이며, axis=1옵션은 '열'을 제거하겠다는 의미다. default는 0이며, 행을 제거한다.
# inplace 옵션을 True로 주면, 변경된 값으로 변수를 다시 저장해준다. 
# 만약 해당 변수가 수정되면 안되는 경우 inplace=False 옵션을 주거나 inplace 옵션을 비워두면 된다. default는 False다.
df_copy.drop('TSLA', axis=1, inplace=True)
df_copy
````

````python
# 다시 df_close를 확인해보면 원본이 수정되지 않았다는 것을 알 수 있다.
df_close
````

### 2-2. 행 인덱싱

````python
# 행 인덱싱은 다음과 같이 할 수 있다.
# 처음부터 5번째 행까지 추출하는 코드다.
df_close[:5]
````

````python
# 마지막 5개 행만 추출하고 싶다면, 다음과 같이 작성한다.
df_close[-5:]
````

````python
# 마찬가지로 마지막 5개 행도 tail() 함수를 사용해 추출할 수 있다.
df_close.tail()
````

### 2-3. 행, 열 동시 인덱싱

````python
# 두 방식을 같이 사용하면 loc행과 열을 함께 인덱싱할 수도 있다.
df_close[:10][['SPY','TSLA']]
````

````python
# 순서는 상관없다.
df_close[['SPY','TSLA']][:10]
````

````python
# 자체 함수를 사용할 수도 있다.
# 함수는 loc()와 iloc()가 있으며, 
# loc는 location의 약자다. 
# 아래 코드는 첫 행부터 2020-01-13 인덱스가 위치한 행까지, 첫 열부터 'SPY' 컬럼이 위치한 열까지 추출한다.
df_close.loc[:'2020-01-13', :'SPY']
````

````python
# 만약 '~까지'가 아니라 특정 행, 특정 열을 지정하고 싶다면 다음과 같이 작성할 수 있다.
df_close.loc['2020-01-13', 'SPY']
````

````python
# 물론 여러 컬럼, 여러 행을 동시에 추출하는 것도 가능하다.
# 이 때는 대괄호를 한번 더 씌워준다.
df_close.loc[['2020-01-13','2019-12-31'], ['SPY','QQQ']]
````

````python
# 특정 인덱스와 컬럼 이름 대신 숫자로 위치만 넘겨줄 수도 있다.
# 이 때는 iloc(integer location)을 사용한다.
# 다음 코드는 첫 행부터 10개 행까지, 첫 열부터 2번째 열까지 추출한다.
df_close.iloc[:10, :2]
````
